```python
dataset = data.values
# split data into X and y
X = dataset[:,0:-1]
X = X.astype(str)
Y = dataset[:,-1]
# encode string input values as integers
enc = OneHotEncoder(handle_unknown='ignore')
enc.fit(X)
enc.categories_
encoded_x = pd.DataFrame(enc.transform(X).toarray())
# encode string class values as integers
label_encoder = LabelEncoder()
encoded_y = label_encoder.fit_transform(Y)

# split data into train and test sets
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(encoded_x, encoded_y, test_size=test_size, random_state=seed)
```

```python
# fit model on training data
model = XGBClassifier(learning_rate=0.001, n_estimators=50, max_depth=4,use_label_encoder=False,
                        min_child_weight=3, gamma=1, subsample=0.6, colsample_bytree=0.9,
                        objective='multi:softmax',  num_class=4, nthread=1, seed=4)
model.fit(X_train, y_train)
print(model)
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
